```python
import folium
import geopandas as gpd
import pandas as pd
import geodatasets
import matplotlib.pyplot as plt
```

```python
pd.set_option("display.max_rows", 20)
pd.set_option("display.max_columns", None)
```

```python
geodatasets.data.geoda.nepal
```

```python
nepal =gpd.read_file(geodatasets.get_path("geoda_nepal"))
```

```python
nepal.info()
```

```python
nepal.head()
```

```python
nepal.plot(figsize=(6,6))
plt.show()
```

```python
nepal.crs
```

```python
m = folium.Map(location=[28.38, 84.14], zoom_start=8)
for _, r in nepal.iterrows():
    sim_geo = gpd.GeoSeries(r["geometry"]).simplify(tolerance=0.001)
    geo_j = sim_geo.to_json()
    geo_j = folium.GeoJson(data=geo_j, style_function=lambda x: {"fillColor": "orange"})
    folium.Popup(r["district"]).add_to(geo_j)
    geo_j.add_to(m)

m
```
